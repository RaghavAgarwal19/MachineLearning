```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')
print(data.shape)
data.head
```

```python
X = data[0].values
Y = data[1].values
```

```python
X = data['YearsExperience'].values
Y = data['Salary'].values
```

```python
X = data['YearsExperience'].values
Y = data['Salary'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)
numer = 0
denom = 0
for i in range(n):
    numer+=(X[i] - mean_x) * (Y[i] - mean_y)
    denom+=(X[i] - mean_x) ** 2
b1 = numer/denom
b0 = mean_y - (b1*mean_x)

print(b1, b0)
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)
numer = 0
denom = 0
for i in range(n):
    numer+=(X[i] - mean_x) * (Y[i] - mean_y)
    denom+=(X[i] - mean_x) ** 2
b1 = numer/denom
b0 = mean_y - (b1*mean_x)

print(b1, b0)
```

```python
max_x = np.max(X) + 10
min_x = np.min(X) - 10

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, color='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
